```python
links = set()
with open("./../url-extractor/discovered_urls.txt") as f:
    for i in f.readlines():
        links.add(i.strip('\n'))
links = list(links)
```

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

loader = WebBaseLoader(web_paths=(links[:50]), bs_kwargs={"parse_only": bs4.SoupStrainer("body")}, header_template=headers)
```

```python
docs = loader.load()
docs
```
